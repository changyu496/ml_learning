```python
# 第一天：Numpy基础练习
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 创建数组
arr1 = np.array([1,2,3,4,5])
arr2 = np.zeros((3, 3))  
arr3 = np.ones((2, 4))   
arr4 = np.random.random((3,3))  
```

```python
print(f"一维数组:\n{arr1}")
print(f"全零数组:\n{arr2}")
print(f"全一数组:\n{arr3}")
print(f"随机数组:\n{arr4}")
```

```python
#2. 数组操作
print(f"数组形状: {arr1.shape}")
print(f"数组维度: {arr1.ndim}")
print(f"数组大小: {arr1.size}")
print(f"数组类型: {arr1.dtype}")
```

```python
# 数组运算
print("\n=== 数组运算 ===")
arr5 = np.array([10, 20, 30, 40, 50])
print(f"数组相加: {arr1 + arr5}")
print(f"数组相乘: {arr1 * arr5}")
print(f"数组求和: {np.sum(arr1)}")
print(f"数组平均值: {np.mean(arr1)}")
```

```python
# 4.矩阵运算
print("\n=== 矩阵运算 ===")
matrix1 = np.array([[1,2],[3,4]])
matrix2 = np.array([[5,6],[7,8]])
print(f"矩阵1:\n {matrix1}")
print(f"矩阵2:\n {matrix2}")
print(f"矩阵相乘: \n{np.dot(matrix1, matrix2)}")
```

```python
# 5. 数据可视化
print("\n=== 数据可视化 ===")
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x,y,"b-",linewidth=2,label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title("我的第一个数据可视化")
plt.grid(True)
plt.legend()
plt.show()

print("练习完成！")
```
