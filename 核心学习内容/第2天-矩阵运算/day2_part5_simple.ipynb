## 1ï¸âƒ£ ç‰¹å¾å€¼ = é‡è¦ç¨‹åº¦

### ğŸ¤” æœ€ç®€å•çš„ç†è§£
- **ç‰¹å¾å€¼**ï¼šä¸€ä¸ªæ•°å­—ï¼Œå‘Šè¯‰ä½ æŸä¸ªæ–¹å‘æœ‰å¤šé‡è¦
- **æ•°å€¼è¶Šå¤§** = è¿™ä¸ªæ–¹å‘è¶Šé‡è¦
- **ç”¨é€”**ï¼šå¸®åŠ©æˆ‘ä»¬æ‰¾åˆ°æ•°æ®ä¸­æœ€é‡è¦çš„ä¿¡æ¯

### ğŸ› ï¸ å®é™…åº”ç”¨
- **PCAé™ç»´**ï¼šä¿ç•™é‡è¦çš„æ–¹å‘ï¼Œä¸¢æ‰ä¸é‡è¦çš„
- **å›¾åƒå‹ç¼©**ï¼šä¿ç•™é‡è¦çš„åƒç´ æ¨¡å¼
- **æ¨èç³»ç»Ÿ**ï¼šæ‰¾åˆ°ç”¨æˆ·è¡Œä¸ºçš„ä¸»è¦æ¨¡å¼

### ğŸ’¡ è®°ä½è¿™å¥è¯
> "ç‰¹å¾å€¼å¤§çš„æ–¹å‘åŒ…å«æ›´å¤šä¿¡æ¯ï¼Œå€¼å¾—ä¿ç•™"

**ä¸éœ€è¦ç†è§£æ•°å­¦å…¬å¼ï¼Œåªè¦çŸ¥é“è¿™ä¸ªç”¨é€”å°±å¤Ÿäº†ï¼**


```python
# 1ï¸âƒ£ ç‰¹å¾å€¼ç¤ºä¾‹ï¼šçœ‹çœ‹æ•°å­—å°±å¤Ÿäº†
print("=== ç‰¹å¾å€¼ç¤ºä¾‹ ===")

# åˆ›å»ºä¸€ä¸ªç®€å•çš„çŸ©é˜µ
A = np.array([[3, 1], 
              [0, 2]])

print(f"çŸ©é˜µA:")
print(A)

# è®¡ç®—ç‰¹å¾å€¼ï¼ˆä¸éœ€è¦ç†è§£æ€ä¹ˆè®¡ç®—çš„ï¼‰
eigenvalues, eigenvectors = np.linalg.eig(A)


print(f"\nç‰¹å¾å€¼: {eigenvalues}")
print(f"æ’åºå: {sorted(eigenvalues, reverse=True)}")

# å…³é”®ç†è§£
max_eigenval = max(eigenvalues)
min_eigenval = min(eigenvalues)

print(f"\nğŸ’¡ å…³é”®ç†è§£:")
print(f"æœ€å¤§ç‰¹å¾å€¼: {max_eigenval:.1f}")
print(f"æœ€å°ç‰¹å¾å€¼: {min_eigenval:.1f}")
print(f"é‡è¦æ€§æ¯”ä¾‹: {max_eigenval/min_eigenval:.1f}:1")
print(f"\nç»“è®º: ç¬¬ä¸€ä¸ªæ–¹å‘æ¯”ç¬¬äºŒä¸ªæ–¹å‘é‡è¦ {max_eigenval/min_eigenval:.1f} å€")

print(f"\nğŸ¯ ä½ åªéœ€è¦è®°ä½:")
print(f"- np.linalg.eig() å¯ä»¥è®¡ç®—ç‰¹å¾å€¼")
print(f"- æ•°å€¼å¤§çš„æ–¹å‘æ›´é‡è¦")
print(f"- è¿™å°±æ˜¯PCAçš„æ ¸å¿ƒåŸç†")
```

## 2ï¸âƒ£ è¡Œåˆ—å¼ = èƒ½å¦æ±‚é€†

### ğŸ¤” æœ€ç®€å•çš„ç†è§£
- **è¡Œåˆ—å¼**ï¼šä¸€ä¸ªæ•°å­—ï¼Œå‘Šè¯‰ä½ çŸ©é˜µèƒ½ä¸èƒ½æ±‚é€†
- **det = 0** â†’ ä¸èƒ½æ±‚é€†
- **det â‰  0** â†’ å¯ä»¥æ±‚é€†

### ğŸ› ï¸ å®é™…åº”ç”¨
- **æ–¹ç¨‹ç»„æ±‚è§£**ï¼šåˆ¤æ–­æ˜¯å¦æœ‰å”¯ä¸€è§£
- **ç‰¹å¾å€¼è®¡ç®—**ï¼šdet(A - Î»I) = 0 æ‰¾ç‰¹å¾å€¼
- **ç³»ç»Ÿç¨³å®šæ€§**ï¼šåˆ¤æ–­ç³»ç»Ÿæ˜¯å¦ç¨³å®š

### ğŸ’¡ è®°ä½è¿™å¥è¯
> "è¡Œåˆ—å¼ä¸º0ï¼ŒçŸ©é˜µå°±ä¸å¯é€†"

**è¿™æ˜¯ä¸ªåˆ¤æ–­å·¥å…·ï¼Œä¸éœ€è¦ç†è§£å‡ ä½•æ„ä¹‰ï¼**


```python
# 2ï¸âƒ£ è¡Œåˆ—å¼ç¤ºä¾‹ï¼šåˆ¤æ–­èƒ½å¦æ±‚é€†
print("=== è¡Œåˆ—å¼ç¤ºä¾‹ ===")

# æµ‹è¯•å‡ ä¸ªä¸åŒçš„çŸ©é˜µ
matrices = [
    ("å¯é€†çŸ©é˜µ", np.array([[3, 1], [2, 4]])),
    ("ä¸å¯é€†çŸ©é˜µ", np.array([[2, 1], [4, 2]])),  # ç¬¬äºŒè¡Œæ˜¯ç¬¬ä¸€è¡Œçš„2å€
    ("å•ä½çŸ©é˜µ", np.array([[1, 0], [0, 1]]))
]

for name, matrix in matrices:
    print(f"\n--- {name} ---")
    print(f"çŸ©é˜µ:\n{matrix}")
    
    # è®¡ç®—è¡Œåˆ—å¼
    det = np.linalg.det(matrix)
    print(f"è¡Œåˆ—å¼: {det:.3f}")
    
    # åˆ¤æ–­èƒ½å¦æ±‚é€†
    if abs(det) < 1e-10:  # è¿‘ä¼¼ä¸º0
        print("âŒ ä¸å¯é€†ï¼ˆè¡Œåˆ—å¼â‰ˆ0ï¼‰")
        print("åŸå› ï¼šçŸ©é˜µçš„è¡Œæˆ–åˆ—çº¿æ€§ç›¸å…³")
    else:
        print("âœ… å¯é€†ï¼ˆè¡Œåˆ—å¼â‰ 0ï¼‰")
        print("é€†çŸ©é˜µå­˜åœ¨")

print(f"\nğŸ¯ ä½ åªéœ€è¦è®°ä½:")
print(f"- np.linalg.det() è®¡ç®—è¡Œåˆ—å¼")
print(f"- det = 0 â†’ ä¸å¯é€†")
print(f"- det â‰  0 â†’ å¯é€†")
print(f"- è¿™ä¸ªåˆ¤æ–­åœ¨æœºå™¨å­¦ä¹ ä¸­å¾ˆæœ‰ç”¨")
```

```python
# 3ï¸âƒ£ çŸ©é˜µä¹˜æ³•ç¤ºä¾‹ï¼šæ•°æ®å˜æ¢
print("=== çŸ©é˜µä¹˜æ³•ç¤ºä¾‹ ===")

# åŸå§‹æ•°æ®ï¼š3ä¸ªæ•°æ®ç‚¹ï¼Œæ¯ä¸ªæœ‰2ä¸ªç‰¹å¾
data = np.array([[1, 2],
                 [3, 4], 
                 [5, 6]])

print(f"åŸå§‹æ•°æ®ï¼ˆ3ä¸ªç‚¹ï¼Œ2ä¸ªç‰¹å¾ï¼‰:")
print(data)
print(f"æ•°æ®å½¢çŠ¶: {data.shape}")

# å˜æ¢çŸ©é˜µï¼šå°†2ä¸ªç‰¹å¾å˜æ¢ä¸º3ä¸ªç‰¹å¾
transform_matrix = np.array([[1, 0, 1],    # æ–°ç‰¹å¾1 = åŸç‰¹å¾1
                            [0, 1, 1]])    # æ–°ç‰¹å¾2 = åŸç‰¹å¾2
                                           # æ–°ç‰¹å¾3 = åŸç‰¹å¾1 + åŸç‰¹å¾2

print(f"\nå˜æ¢çŸ©é˜µï¼ˆ2â†’3ç‰¹å¾ï¼‰:")
print(transform_matrix)
print(f"å˜æ¢çŸ©é˜µå½¢çŠ¶: {transform_matrix.shape}")

# æ‰§è¡Œå˜æ¢ï¼šæ•°æ® @ å˜æ¢çŸ©é˜µ
transformed_data = data @ transform_matrix

print(f"\nå˜æ¢åçš„æ•°æ®ï¼ˆ3ä¸ªç‚¹ï¼Œ3ä¸ªç‰¹å¾ï¼‰:")
print(transformed_data)
print(f"æ–°æ•°æ®å½¢çŠ¶: {transformed_data.shape}")

# è§£é‡Šå˜æ¢ç»“æœ
print(f"\nğŸ’¡ å˜æ¢è§£é‡Š:")
for i in range(len(data)):
    original = data[i]
    new = transformed_data[i]
    print(f"ç‚¹{i+1}: {original} â†’ {new}")

print(f"\nğŸ¯ ä½ åªéœ€è¦è®°ä½:")
print(f"- ç”¨ @ æˆ– np.dot() åšçŸ©é˜µä¹˜æ³•")
print(f"- å½¢çŠ¶è¦åŒ¹é…ï¼š(m,n) @ (n,k) = (m,k)")
print(f"- è¿™æ˜¯æ•°æ®å˜æ¢çš„åŸºç¡€å·¥å…·")
print(f"- æœºå™¨å­¦ä¹ ä¸­åˆ°å¤„éƒ½åœ¨ç”¨")
```
