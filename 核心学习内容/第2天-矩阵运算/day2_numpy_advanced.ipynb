3. å¦‚æœæŸä¸ªç»´åº¦å¤§å°ä¸º1ï¼Œé‚£ä¹ˆè¯¥ç»´åº¦ä¼šè¢«"æ‹‰ä¼¸"ä»¥åŒ¹é…å¦ä¸€ä¸ªæ•°ç»„

```python
# å¹¿æ’­æœºåˆ¶æ¼”ç¤º
print("=== å¹¿æ’­æœºåˆ¶æ¼”ç¤º ===")

# ç¤ºä¾‹1ï¼šçŸ©é˜µä¸å‘é‡
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])
vector_b = np.array([10, 20, 30])

print(f"çŸ©é˜µAå½¢çŠ¶: {matrix_a.shape}")
print(f"çŸ©é˜µA:\n{matrix_a}")
print(f"å‘é‡Bå½¢çŠ¶: {vector_b.shape}")
print(f"å‘é‡B: {vector_b}")

# å¹¿æ’­ç›¸åŠ 
result_add = matrix_a + vector_b
print(f"\nå¹¿æ’­ç›¸åŠ ç»“æœ:\n{result_add}")

# å¹¿æ’­ç›¸ä¹˜
result_mul = matrix_a * vector_b
print(f"\nå¹¿æ’­ç›¸ä¹˜ç»“æœ:\n{result_mul}")
```

```python
# ç¤ºä¾‹3ï¼šå¹¿æ’­åœ¨å®é™…åº”ç”¨ä¸­çš„ä¾‹å­
print("\n=== å®é™…åº”ç”¨ç¤ºä¾‹ ===")
# å‡è®¾æˆ‘ä»¬æœ‰ä¸€ä¸ªå›¾åƒæ•°æ®ï¼ˆç®€åŒ–ç‰ˆï¼‰
image = np.random.randint(0, 256, (3, 4, 4))  # 3é€šé“ï¼Œ4x4å›¾åƒ
print(f"å›¾åƒæ•°æ®å½¢çŠ¶: {image.shape}")

# æˆ‘ä»¬æƒ³è¦å¯¹æ¯ä¸ªé€šé“åº”ç”¨ä¸åŒçš„æƒé‡
weights = np.array([0.3, 0.6, 0.1])  # RGBæƒé‡
print(f"æƒé‡å½¢çŠ¶: {weights.shape}")

# ä½¿ç”¨å¹¿æ’­åº”ç”¨æƒé‡
weighted_image = image * weights[:, np.newaxis, np.newaxis]
print(f"åŠ æƒåå›¾åƒå½¢çŠ¶: {weighted_image.shape}")
```

```python
# ç»ƒä¹ ä»»åŠ¡
print("\n=== ç»ƒä¹ ä»»åŠ¡ ===")
# 1. åˆ›å»ºä¸€ä¸ª5x3çš„çŸ©é˜µå’Œä¸€ä¸ªé•¿åº¦ä¸º3çš„å‘é‡ï¼Œè¿›è¡Œå¹¿æ’­ç›¸åŠ 
matrix1 = np.arange(15).reshape(5,3)
print(f"5x3çŸ©é˜µ:\n{matrix1}")
vector1 = np.arange(3)
print(f"é•¿åº¦ä¸º3çš„å‘é‡:\n{vector1}")
print(f"5x3çŸ©é˜µ å’Œé•¿åº¦ä¸º3çš„å‘é‡ å¹¿æ’­ç›¸åŠ :\n{matrix1 + vector1}")

# 2. åˆ›å»ºä¸€ä¸ª3x1çš„çŸ©é˜µå’Œä¸€ä¸ªé•¿åº¦ä¸º4çš„å‘é‡ï¼Œè¿›è¡Œå¹¿æ’­ç›¸ä¹˜
matrix2 = np.arange(3).reshape(3,1)
vector2 = np.arange(4)
print(f"3x1çŸ©é˜µ:\n{matrix2}")
print(f"é•¿åº¦ä¸º4çš„å‘é‡:\n{vector2}")
print(f"3x1çŸ©é˜µ å’Œé•¿åº¦ä¸º4çš„å‘é‡ å¹¿æ’­ç›¸ä¹˜:\n{matrix2 * vector2}")

print("\nâœ… å¹¿æ’­æœºåˆ¶ç»ƒä¹ å®Œæˆï¼")
```

## 3. æ•°ç»„å½¢çŠ¶æ“ä½œ

### ä¸»è¦å‡½æ•°ï¼š

- `transpose()` æˆ– `.T`: è½¬ç½®
- `flatten()`: å±•å¹³æ•°ç»„
- `ravel()`: å±•å¹³æ•°ç»„ï¼ˆè¿”å›è§†å›¾ï¼‰
- `vstack()` å’Œ `hstack()`: æ•°ç»„æ‹¼æ¥

```python
# æ•°ç»„å½¢çŠ¶æ“ä½œæ¼”ç¤º
print("=== æ•°ç»„å½¢çŠ¶æ“ä½œ ===")

# åˆ›å»ºåŸå§‹æ•°ç»„
original = np.arange(12)
print(f"åŸå§‹æ•°ç»„: {original}")
print(f"åŸå§‹å½¢çŠ¶: {original.shape}")
```

```python
# 1. é‡å¡‘å½¢çŠ¶
reshaped_2d = original.reshape(3, 4)
print(f"\né‡å¡‘ä¸º3x4:\n{reshaped_2d}")

reshaped_3d = original.reshape(2, 2, 3)
print(f"\né‡å¡‘ä¸º2x2x3:\n{reshaped_3d}")
```

```python
# 2. è½¬ç½®
transposed = reshaped_2d.T
print(f"\nè½¬ç½®å:\n{transposed}")
```

```python
# 3. å±•å¹³
flattened = reshaped_2d.flatten()
print(f"\nå±•å¹³å: {flattened}")

# ravelå’Œflattençš„åŒºåˆ«
raveled = reshaped_2d.ravel()
print(f"ravelå: {raveled}")

# ä¿®æ”¹ravelçš„ç»“æœä¼šå½±å“åŸæ•°ç»„
raveled[0] = 999
print(f"ä¿®æ”¹ravelåï¼ŒåŸæ•°ç»„:\n{reshaped_2d}")

# é‡æ–°åˆ›å»ºæ•°ç»„ç”¨äºåç»­æ¼”ç¤º
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
print(f"\næ•°ç»„1:\n{arr1}")
print(f"æ•°ç»„2:\n{arr2}")
```

## 4. ç»Ÿè®¡å’Œèšåˆå‡½æ•°

### å¸¸ç”¨ç»Ÿè®¡å‡½æ•°ï¼š
- `np.sum()`: æ±‚å’Œ
- `np.mean()`: å¹³å‡å€¼
- `np.median()`: ä¸­ä½æ•°
- `np.std()`: æ ‡å‡†å·®
- `np.var()`: æ–¹å·®
- `np.max()`, `np.min()`: æœ€å¤§å€¼å’Œæœ€å°å€¼
- `np.argmax()`, `np.argmin()`: æœ€å¤§å€¼å’Œæœ€å°å€¼çš„ç´¢å¼•
- `np.percentile()`: ç™¾åˆ†ä½æ•°


```python
# ç»Ÿè®¡å’Œèšåˆå‡½æ•°æ¼”ç¤º
print("=== ç»Ÿè®¡å’Œèšåˆå‡½æ•° ===")

# åˆ›å»ºæµ‹è¯•æ•°æ®
np.random.seed(42)
data = np.random.randint(1, 100, size=(5, 4))
print(f"æµ‹è¯•æ•°æ®:\n{data}")
print(f"æ•°æ®å½¢çŠ¶: {data.shape}")

# åŸºæœ¬ç»Ÿè®¡
print(f"\n=== åŸºæœ¬ç»Ÿè®¡ ===")
print(f"æ€»å’Œ: {np.sum(data)}")
print(f"å¹³å‡å€¼: {np.mean(data):.2f}")
print(f"ä¸­ä½æ•°: {np.median(data)}")
print(f"æœ€å¤§å€¼: {np.max(data)}")
print(f"æœ€å°å€¼: {np.min(data)}")
print(f"æ ‡å‡†å·®: {np.std(data):.2f}")
print(f"æ–¹å·®: {np.var(data):.2f}")

# æœ€å¤§å€¼å’Œæœ€å°å€¼çš„ä½ç½®
print(f"\n=== ä½ç½®ä¿¡æ¯ ===")
print(f"æœ€å¤§å€¼ä½ç½®: {np.argmax(data)}")
print(f"æœ€å°å€¼ä½ç½®: {np.argmin(data)}")

# å°†ä¸€ç»´ç´¢å¼•è½¬æ¢ä¸ºäºŒç»´ç´¢å¼•
max_pos = np.unravel_index(np.argmax(data), data.shape)
min_pos = np.unravel_index(np.argmin(data), data.shape)
print(f"æœ€å¤§å€¼ä½ç½®(è¡Œ,åˆ—): {max_pos}")
print(f"æœ€å°å€¼ä½ç½®(è¡Œ,åˆ—): {min_pos}")

# æŒ‰è½´ç»Ÿè®¡
print(f"\n=== æŒ‰è½´ç»Ÿè®¡ ===")
print(f"æŒ‰è¡Œæ±‚å’Œ (axis=1): {np.sum(data, axis=1)}")
print(f"æŒ‰åˆ—æ±‚å’Œ (axis=0): {np.sum(data, axis=0)}")
print(f"æŒ‰è¡Œæ±‚å¹³å‡ (axis=1): {np.mean(data, axis=1)}")
print(f"æŒ‰åˆ—æ±‚å¹³å‡ (axis=0): {np.mean(data, axis=0)}")

# ç´¯ç§¯ç»Ÿè®¡
print(f"\n=== ç´¯ç§¯ç»Ÿè®¡ ===")
print(f"ç´¯ç§¯å’Œ: {np.cumsum(data.flatten())}")
print(f"ç´¯ç§¯ä¹˜ç§¯: {np.cumprod(data.flatten())}")

# ç™¾åˆ†ä½æ•°
print(f"\n=== ç™¾åˆ†ä½æ•° ===")
print(f"25%åˆ†ä½æ•°: {np.percentile(data, 25)}")
print(f"50%åˆ†ä½æ•°: {np.percentile(data, 50)}")
print(f"75%åˆ†ä½æ•°: {np.percentile(data, 75)}")

# æ¡ä»¶ç»Ÿè®¡
print(f"\n=== æ¡ä»¶ç»Ÿè®¡ ===")
print(f"å¤§äº50çš„å…ƒç´ æ•°é‡: {np.sum(data > 50)}")
print(f"å¤§äº50çš„å…ƒç´ æ¯”ä¾‹: {np.mean(data > 50):.2%}")
print(f"æ¯è¡Œä¸­å¤§äº50çš„å…ƒç´ æ•°é‡: {np.sum(data > 50, axis=1)}")

# è‡ªå®šä¹‰ç»Ÿè®¡å‡½æ•°
def custom_range(arr):
    """è‡ªå®šä¹‰å‡½æ•°ï¼šè®¡ç®—æ•°ç»„çš„èŒƒå›´ï¼ˆæœ€å¤§å€¼-æœ€å°å€¼ï¼‰"""
    return np.max(arr) - np.min(arr)

print(f"\n=== è‡ªå®šä¹‰ç»Ÿè®¡ ===")
print(f"æ•°æ®èŒƒå›´: {custom_range(data)}")
print(f"æ¯è¡Œçš„èŒƒå›´: {[custom_range(row) for row in data]}")

# ç»ƒä¹ ä»»åŠ¡
print("\n=== ç»ƒä¹ ä»»åŠ¡ ===")
# 1. åˆ›å»ºä¸€ä¸ªæ­£æ€åˆ†å¸ƒçš„æ•°æ®
normal_data = np.random.normal(50,15,100)
print(f"å‡å€¼: {np.mean(normal_data)}")
print(f"æ ‡å‡†å·®: {np.std(normal_data)}")
print(f"æœ€å¤§å€¼: {np.max(normal_data)}")
print(f"æœ€å°å€¼: {np.min(normal_data)}")

# 2. åˆ†æä¸€ä¸ªæ›´å¤æ‚çš„æ•°æ®é›†
complex_data = np.random.rand(2,3)
print(f"å¤æ‚æ•°æ®:\n{complex_data}")
print(f"å¤æ‚æ•°æ®å½¢çŠ¶: {complex_data.shape}")
# æ•°å­—è¶Šå°ï¼Œå°±è¶Šé å¤–è¾¹
print(f"å¤æ‚æ•°æ®æŒ‰ axis=0 çš„å¹³å‡å€¼: {np.mean(complex_data,axis=0)}")
print(f"å¤æ‚æ•°æ®æŒ‰ axis=1 çš„å¹³å‡å€¼: {np.mean(complex_data,axis=1)}")


print("\nâœ… ç»Ÿè®¡å’Œèšåˆå‡½æ•°ç»ƒä¹ å®Œæˆï¼")
```

## 5. çŸ©é˜µè¿ç®—å‡½æ•°åº“å®ç°

### å­¦ä¹ ç›®æ ‡ï¼š
- åˆ›å»ºä¸€ä¸ªè‡ªå®šä¹‰çš„çŸ©é˜µè¿ç®—ç±»
- å®ç°å¸¸ç”¨çš„çŸ©é˜µè¿ç®—å‡½æ•°

```python
# åˆ›å»ºçŸ©é˜µè¿ç®—å‡½æ•°åº“
class MatrixOperations:
    """çŸ©é˜µè¿ç®—å‡½æ•°åº“"""
    
    @staticmethod
    def matrix_multiply(A, B):
        """çŸ©é˜µä¹˜æ³•"""
        if A.shape[1] != B.shape[0]:
            raise ValueError(f"çŸ©é˜µç»´åº¦ä¸åŒ¹é…: {A.shape} å’Œ {B.shape}")
        return np.dot(A, B)
    
    @staticmethod
    def matrix_inverse(A):
        """çŸ©é˜µæ±‚é€†"""
        if A.shape[0] != A.shape[1]:
            raise ValueError("å¿…é¡»æ˜¯æ–¹é˜µ")
        det = np.linalg.det(A)
        if abs(det) < 1e-10:
            raise ValueError("çŸ©é˜µä¸å¯é€†ï¼ˆè¡Œåˆ—å¼ä¸º0ï¼‰")
        return np.linalg.inv(A)
    
    @staticmethod
    def matrix_determinant(A):
        """è®¡ç®—è¡Œåˆ—å¼"""
        if A.shape[0] != A.shape[1]:
            raise ValueError("å¿…é¡»æ˜¯æ–¹é˜µ")
        return np.linalg.det(A)
    
    @staticmethod
    def matrix_trace(A):
        """è®¡ç®—çŸ©é˜µçš„è¿¹"""
        if A.shape[0] != A.shape[1]:
            raise ValueError("å¿…é¡»æ˜¯æ–¹é˜µ")
        return np.trace(A)
    
    @staticmethod
    def matrix_eigenvalues(A):
        """è®¡ç®—ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡"""
        if A.shape[0] != A.shape[1]:
            raise ValueError("å¿…é¡»æ˜¯æ–¹é˜µ")
        eigenvalues, eigenvectors = np.linalg.eig(A)
        return eigenvalues, eigenvectors
    
    @staticmethod
    def matrix_rank(A):
        """è®¡ç®—çŸ©é˜µçš„ç§©"""
        return np.linalg.matrix_rank(A)
    
    @staticmethod
    def matrix_norm(A, ord=None):
        """è®¡ç®—çŸ©é˜µèŒƒæ•°"""
        return np.linalg.norm(A, ord=ord)
    
    @staticmethod
    def matrix_condition_number(A):
        """è®¡ç®—çŸ©é˜µæ¡ä»¶æ•°"""
        return np.linalg.cond(A)

# æµ‹è¯•çŸ©é˜µè¿ç®—å‡½æ•°åº“
print("=== çŸ©é˜µè¿ç®—å‡½æ•°åº“æµ‹è¯• ===")

# åˆ›å»ºæµ‹è¯•çŸ©é˜µ
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6], 
              [7, 8]])
C = np.array([[2, 1], 
              [1, 2]])

print(f"çŸ©é˜µA:\n{A}")
print(f"çŸ©é˜µB:\n{B}")
print(f"çŸ©é˜µC:\n{C}")

# æµ‹è¯•çŸ©é˜µè¿ç®—
mat_ops = MatrixOperations()

# çŸ©é˜µä¹˜æ³•
print(f"\n=== çŸ©é˜µä¹˜æ³• ===")
AB = mat_ops.matrix_multiply(A, B)
print(f"A Ã— B:\n{AB}")

# è¡Œåˆ—å¼
print(f"\n=== è¡Œåˆ—å¼ ===")
det_A = mat_ops.matrix_determinant(A)
det_B = mat_ops.matrix_determinant(B)
det_C = mat_ops.matrix_determinant(C)
print(f"det(A) = {det_A}")
print(f"det(B) = {det_B}")
print(f"det(C) = {det_C}")

# çŸ©é˜µçš„è¿¹
print(f"\n=== çŸ©é˜µçš„è¿¹ ===")
trace_A = mat_ops.matrix_trace(A)
trace_B = mat_ops.matrix_trace(B)
print(f"trace(A) = {trace_A}")
print(f"trace(B) = {trace_B}")

# ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡
print(f"\n=== ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡ ===")
eigenvals_A, eigenvecs_A = mat_ops.matrix_eigenvalues(A)
print(f"Açš„ç‰¹å¾å€¼: {eigenvals_A}")
print(f"Açš„ç‰¹å¾å‘é‡:\n{eigenvecs_A}")

# éªŒè¯ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡
print(f"\n=== éªŒè¯ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡ ===")
for i in range(len(eigenvals_A)):
    lambda_i = eigenvals_A[i]
    v_i = eigenvecs_A[:, i]
    Av_i = A @ v_i
    lambda_v_i = lambda_i * v_i
    print(f"Î»{i+1} = {lambda_i:.3f}")
    print(f"Av{i+1} = {Av_i}")
    print(f"Î»{i+1}v{i+1} = {lambda_v_i}")
    print(f"è¯¯å·®: {np.allclose(Av_i, lambda_v_i)}")

# çŸ©é˜µæ±‚é€†
print(f"\n=== çŸ©é˜µæ±‚é€† ===")
try:
    inv_C = mat_ops.matrix_inverse(C)
    print(f"Cçš„é€†çŸ©é˜µ:\n{inv_C}")
    
    # éªŒè¯é€†çŸ©é˜µ
    identity = C @ inv_C
    print(f"C Ã— C^(-1) =\n{identity}")
    print(f"æ˜¯å¦ä¸ºå•ä½çŸ©é˜µ: {np.allclose(identity, np.eye(2))}")
except ValueError as e:
    print(f"é”™è¯¯: {e}")

# çŸ©é˜µç§©
print(f"\n=== çŸ©é˜µç§© ===")
rank_A = mat_ops.matrix_rank(A)
rank_B = mat_ops.matrix_rank(B)
print(f"rank(A) = {rank_A}")
print(f"rank(B) = {rank_B}")

# çŸ©é˜µèŒƒæ•°
print(f"\n=== çŸ©é˜µèŒƒæ•° ===")
norm_A_fro = mat_ops.matrix_norm(A, 'fro')  # FrobeniusèŒƒæ•°
norm_A_2 = mat_ops.matrix_norm(A, 2)        # 2-èŒƒæ•°
print(f"Açš„FrobeniusèŒƒæ•°: {norm_A_fro:.3f}")
print(f"Açš„2-èŒƒæ•°: {norm_A_2:.3f}")

# æ¡ä»¶æ•°
print(f"\n=== æ¡ä»¶æ•° ===")
cond_A = mat_ops.matrix_condition_number(A)
cond_C = mat_ops.matrix_condition_number(C)
print(f"Açš„æ¡ä»¶æ•°: {cond_A:.3f}")
print(f"Cçš„æ¡ä»¶æ•°: {cond_C:.3f}")

# å®é™…åº”ç”¨ç¤ºä¾‹ï¼šè§£çº¿æ€§æ–¹ç¨‹ç»„
print(f"\n=== çº¿æ€§æ–¹ç¨‹ç»„æ±‚è§£ ===")
# è§£æ–¹ç¨‹ç»„ Cx = b
b = np.array([1, 2])
x = np.linalg.solve(C, b)
print(f"æ–¹ç¨‹ç»„ Cx = b")
print(f"C =\n{C}")
print(f"b = {b}")
print(f"è§£ x = {x}")

# éªŒè¯è§£
verification = C @ x
print(f"éªŒè¯ Cx = {verification}")
print(f"æ˜¯å¦æ­£ç¡®: {np.allclose(verification, b)}")

print("\nâœ… çŸ©é˜µè¿ç®—å‡½æ•°åº“æµ‹è¯•å®Œæˆï¼")
```

```python
# å®ç°ç®€å•çš„PCAä¸»æˆåˆ†åˆ†æ
def simple_pca(X, n_components=2):
    """
    ç®€å•çš„PCAå®ç°
    
    å‚æ•°:
    X: è¾“å…¥æ•°æ®çŸ©é˜µ (n_samples, n_features)
    n_components: ä¸»æˆåˆ†æ•°é‡
    
    è¿”å›:
    X_pca: é™ç»´åçš„æ•°æ®
    components: ä¸»æˆåˆ†ï¼ˆç‰¹å¾å‘é‡ï¼‰
    eigenvalues: ç‰¹å¾å€¼
    explained_variance_ratio: è§£é‡Šæ–¹å·®æ¯”ä¾‹
    """
    # æ­¥éª¤1: æ•°æ®ä¸­å¿ƒåŒ–
    X_mean = np.mean(X, axis=0)
    X_centered = X - X_mean
    
    # æ­¥éª¤2: è®¡ç®—åæ–¹å·®çŸ©é˜µ
    cov_matrix = np.cov(X_centered.T)
    
    # æ­¥éª¤3: è®¡ç®—ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # æ­¥éª¤4: æŒ‰ç‰¹å¾å€¼é™åºæ’åˆ—
    idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    
    # æ­¥éª¤5: é€‰æ‹©å‰n_componentsä¸ªä¸»æˆåˆ†
    components = eigenvectors[:, :n_components]
    
    # æ­¥éª¤6: æŠ•å½±æ•°æ®åˆ°æ–°ç©ºé—´
    X_pca = np.dot(X_centered, components)
    
    # è®¡ç®—è§£é‡Šæ–¹å·®æ¯”ä¾‹
    explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
    
    return X_pca, components, eigenvalues, explained_variance_ratio

# æµ‹è¯•PCAç®—æ³•
print("=== PCAç®—æ³•æµ‹è¯• ===")

# åˆ›å»ºå…·æœ‰ç›¸å…³æ€§çš„æµ‹è¯•æ•°æ®
np.random.seed(42)
n_samples = 300
n_features = 4

# ç”ŸæˆåŸå§‹æ•°æ®
X = np.random.randn(n_samples, n_features)

# åœ¨ç‰¹å¾ä¹‹é—´åˆ›å»ºç›¸å…³æ€§
X[:, 1] = X[:, 0] + 0.5 * np.random.randn(n_samples)  # ç‰¹å¾1ä¸ç‰¹å¾0ç›¸å…³
X[:, 2] = -X[:, 0] + 0.3 * np.random.randn(n_samples)  # ç‰¹å¾2ä¸ç‰¹å¾0è´Ÿç›¸å…³
X[:, 3] = 0.2 * X[:, 0] + 0.8 * np.random.randn(n_samples)  # ç‰¹å¾3ä¸ç‰¹å¾0å¼±ç›¸å…³

print(f"åŸå§‹æ•°æ®å½¢çŠ¶: {X.shape}")
print(f"åŸå§‹æ•°æ®å‰5è¡Œ:\n{X[:5]}")

# è®¡ç®—åŸå§‹æ•°æ®çš„åæ–¹å·®çŸ©é˜µ
orig_cov = np.cov(X.T)
print(f"\nåŸå§‹æ•°æ®åæ–¹å·®çŸ©é˜µ:\n{orig_cov}")

# åº”ç”¨PCA
X_pca, components, eigenvalues, explained_variance_ratio = simple_pca(X, n_components=2)

print(f"\nPCAç»“æœ:")
print(f"é™ç»´åæ•°æ®å½¢çŠ¶: {X_pca.shape}")
print(f"ä¸»æˆåˆ†ï¼ˆç‰¹å¾å‘é‡ï¼‰:\n{components}")
print(f"ç‰¹å¾å€¼: {eigenvalues}")
print(f"è§£é‡Šæ–¹å·®æ¯”ä¾‹: {explained_variance_ratio}")
print(f"å‰2ä¸ªä¸»æˆåˆ†ç´¯è®¡è§£é‡Šæ–¹å·®æ¯”ä¾‹: {np.sum(explained_variance_ratio[:2]):.3f}")

# éªŒè¯PCAçš„æ•°å­¦ç‰¹æ€§
print(f"\n=== PCAéªŒè¯ ===")
# 1. ä¸»æˆåˆ†åº”è¯¥æ˜¯æ­£äº¤çš„
dot_product = np.dot(components[:, 0], components[:, 1])
print(f"å‰ä¸¤ä¸ªä¸»æˆåˆ†çš„ç‚¹ç§¯: {dot_product:.6f} (åº”è¯¥æ¥è¿‘0)")

# 2. ä¸»æˆåˆ†åº”è¯¥æ˜¯å•ä½å‘é‡
norm1 = np.linalg.norm(components[:, 0])
norm2 = np.linalg.norm(components[:, 1])
print(f"ç¬¬ä¸€ä¸»æˆåˆ†çš„æ¨¡é•¿: {norm1:.6f} (åº”è¯¥ç­‰äº1)")
print(f"ç¬¬äºŒä¸»æˆåˆ†çš„æ¨¡é•¿: {norm2:.6f} (åº”è¯¥ç­‰äº1)")

# 3. é™ç»´åæ•°æ®çš„å‡å€¼åº”è¯¥ä¸º0
mean_pca = np.mean(X_pca, axis=0)
print(f"é™ç»´åæ•°æ®çš„å‡å€¼: {mean_pca} (åº”è¯¥æ¥è¿‘0)")

# æ•°æ®é‡æ„
print(f"\n=== æ•°æ®é‡æ„ ===")
# ä½¿ç”¨å‰2ä¸ªä¸»æˆåˆ†é‡æ„åŸå§‹æ•°æ®
X_reconstructed = np.dot(X_pca, components.T) + np.mean(X, axis=0)
reconstruction_error = np.mean((X - X_reconstructed)**2)
print(f"é‡æ„è¯¯å·® (MSE): {reconstruction_error:.6f}")

# æ¯”è¾ƒåŸå§‹æ•°æ®å’Œé‡æ„æ•°æ®
print(f"åŸå§‹æ•°æ®å‰3è¡Œ:\n{X[:3]}")
print(f"é‡æ„æ•°æ®å‰3è¡Œ:\n{X_reconstructed[:3]}")
```
