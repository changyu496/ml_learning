```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')

# è®¾ç½®ä¸­æ–‡å­—ä½“
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False

print("ğŸš€ å¼€å§‹ç¬¬3å¤©å­¦ä¹ ï¼")
print("è®°ä½ï¼šå‘é‡å°±æ˜¯æ•°æ®ï¼Œå‘é‡è¿ç®—å°±æ˜¯æ•°æ®å¤„ç†ï¼")
```

## 1ï¸âƒ£ å‘é‡ = æ•°æ®ç‚¹

### ğŸ¤” æœ€ç®€å•çš„ç†è§£
- **å‘é‡**ï¼šå°±æ˜¯ä¸€å †æ•°å­—æ’æˆä¸€è¡Œï¼ˆæˆ–ä¸€åˆ—ï¼‰
- **ä¾‹å­**ï¼šç”¨æˆ·è¯„åˆ† [5, 3, 4, 2, 1] å°±æ˜¯ä¸€ä¸ªå‘é‡
- **æ„ä¹‰**ï¼šæ¯ä¸ªæ•°å­—ä»£è¡¨ä¸€ä¸ªç‰¹å¾æˆ–å±æ€§

### ğŸ› ï¸ å®é™…åº”ç”¨
- **ç”¨æˆ·ç”»åƒ**ï¼š[å¹´é¾„, æ”¶å…¥, è´­ä¹°é¢‘æ¬¡, æ»¡æ„åº¦]
- **å•†å“ç‰¹å¾**ï¼š[ä»·æ ¼, è¯„åˆ†, é”€é‡, åº“å­˜]
- **æ–‡æœ¬è¡¨ç¤º**ï¼š[è¯æ±‡1å‡ºç°æ¬¡æ•°, è¯æ±‡2å‡ºç°æ¬¡æ•°, ...]

### ğŸ’¡ æ ¸å¿ƒç†å¿µ
> "å‘é‡ä¸æ˜¯æ•°å­¦æ¦‚å¿µï¼Œæ˜¯æ•°æ®è¡¨ç¤ºæ–¹æ³•"


```python
# 1ï¸âƒ£ å‘é‡ç¤ºä¾‹ï¼šç”¨æˆ·è¯„åˆ†æ•°æ®
print("=== å‘é‡å°±æ˜¯æ•°æ® ===")

# åˆ›å»ºç”¨æˆ·è¯„åˆ†å‘é‡ï¼ˆå¯¹5éƒ¨ç”µå½±çš„è¯„åˆ†ï¼‰
user_A = np.array([5, 3, 4, 2, 1])  # ç”¨æˆ·Açš„è¯„åˆ†
user_B = np.array([4, 3, 5, 2, 2])  # ç”¨æˆ·Bçš„è¯„åˆ†
user_C = np.array([1, 2, 1, 4, 5])  # ç”¨æˆ·Cçš„è¯„åˆ†

print(f"ç”¨æˆ·Açš„è¯„åˆ†å‘é‡: {user_A}")
print(f"ç”¨æˆ·Bçš„è¯„åˆ†å‘é‡: {user_B}")
print(f"ç”¨æˆ·Cçš„è¯„åˆ†å‘é‡: {user_C}")

# å‘é‡çš„åŸºæœ¬ä¿¡æ¯
print(f"\nå‘é‡çš„ç»´åº¦ï¼ˆç‰¹å¾æ•°é‡ï¼‰: {len(user_A)}")
print(f"å‘é‡çš„å½¢çŠ¶: {user_A.shape}")
print(f"å‘é‡çš„ç±»å‹: {type(user_A)}")

# ç”µå½±åç§°ï¼ˆæ–¹ä¾¿ç†è§£ï¼‰
movies = ['åŠ¨ä½œç‰‡', 'å–œå‰§ç‰‡', 'ç§‘å¹»ç‰‡', 'çˆ±æƒ…ç‰‡', 'ææ€–ç‰‡']
print(f"\nğŸ’¡ å…·ä½“å«ä¹‰:")
for i, movie in enumerate(movies):
    print(f"  {movie}: ç”¨æˆ·Aè¯„åˆ†{user_A[i]}, ç”¨æˆ·Bè¯„åˆ†{user_B[i]}, ç”¨æˆ·Cè¯„åˆ†{user_C[i]}")

print(f"\nğŸ¯ å…³é”®ç†è§£:")
print(f"- å‘é‡ = æ•°æ®ç‚¹")
print(f"- æ¯ä¸ªæ•°å­— = ä¸€ä¸ªç‰¹å¾")
print(f"- å‘é‡é•¿åº¦ = ç‰¹å¾æ•°é‡")
print(f"- è¿™å°±æ˜¯æœºå™¨å­¦ä¹ çš„æ•°æ®è¡¨ç¤ºæ–¹æ³•ï¼")
```

## 2ï¸âƒ£ å‘é‡è¿ç®— = æ•°æ®æ“ä½œ

### ğŸ¤” æœ€ç®€å•çš„ç†è§£
- **å‘é‡åŠ æ³•**ï¼šå¯¹åº”ä½ç½®ç›¸åŠ 
- **å‘é‡å‡æ³•**ï¼šå¯¹åº”ä½ç½®ç›¸å‡
- **ç‚¹ç§¯**ï¼šå¯¹åº”ä½ç½®ç›¸ä¹˜å†æ±‚å’Œï¼ˆæœ€é‡è¦ï¼ï¼‰
- **å‘é‡é•¿åº¦**ï¼šè¡¡é‡å‘é‡çš„å¤§å°

### ğŸ› ï¸ å®é™…åº”ç”¨
- **ç›¸ä¼¼åº¦è®¡ç®—**ï¼šç”¨ç‚¹ç§¯åˆ¤æ–­ä¸¤ä¸ªç”¨æˆ·æ˜¯å¦ç›¸ä¼¼
- **æ¨èç³»ç»Ÿ**ï¼šæ‰¾åˆ°ç›¸ä¼¼ç”¨æˆ·ï¼Œæ¨èä»–ä»¬å–œæ¬¢çš„å•†å“
- **åˆ†ç±»ç®—æ³•**ï¼šè®¡ç®—æ•°æ®ç‚¹ä¹‹é—´çš„è·ç¦»

### ğŸ’¡ æ ¸å¿ƒç†å¿µ
> "å‘é‡è¿ç®—ä¸æ˜¯æ•°å­¦æ¸¸æˆï¼Œæ˜¯æ•°æ®åˆ†æå·¥å…·"


```python
# 2ï¸âƒ£ å‘é‡è¿ç®—ç¤ºä¾‹
print("=== å‘é‡è¿ç®— = æ•°æ®æ“ä½œ ===")

# ä½¿ç”¨å‰é¢çš„ç”¨æˆ·è¯„åˆ†æ•°æ®
print(f"ç”¨æˆ·A: {user_A}")
print(f"ç”¨æˆ·B: {user_B}")

# 1. å‘é‡åŠ æ³•ï¼šå¹³å‡è¯„åˆ†
avg_rating = (user_A + user_B) / 2
print(f"\n1ï¸âƒ£ å‘é‡åŠ æ³• - å¹³å‡è¯„åˆ†:")
print(f"(ç”¨æˆ·A + ç”¨æˆ·B) / 2 = {avg_rating}")

# 2. å‘é‡å‡æ³•ï¼šè¯„åˆ†å·®å¼‚
rating_diff = user_A - user_B
print(f"\n2ï¸âƒ£ å‘é‡å‡æ³• - è¯„åˆ†å·®å¼‚:")
print(f"ç”¨æˆ·A - ç”¨æˆ·B = {rating_diff}")
print(f"æ­£æ•°è¡¨ç¤ºAæ¯”Bè¯„åˆ†é«˜ï¼Œè´Ÿæ•°è¡¨ç¤ºAæ¯”Bè¯„åˆ†ä½")

# 3. ç‚¹ç§¯ï¼šç›¸ä¼¼åº¦æŒ‡æ ‡ï¼ˆæœ€é‡è¦ï¼ï¼‰
dot_product = np.dot(user_A, user_B)
print(f"\n3ï¸âƒ£ ç‚¹ç§¯ - ç›¸ä¼¼åº¦æŒ‡æ ‡:")
print(f"ç”¨æˆ·A Â· ç”¨æˆ·B = {dot_product}")
print(f"è®¡ç®—è¿‡ç¨‹: {user_A[0]}Ã—{user_B[0]} + {user_A[1]}Ã—{user_B[1]} + {user_A[2]}Ã—{user_B[2]} + {user_A[3]}Ã—{user_B[3]} + {user_A[4]}Ã—{user_B[4]}")
print(f"         = {user_A[0]*user_B[0]} + {user_A[1]*user_B[1]} + {user_A[2]*user_B[2]} + {user_A[3]*user_B[3]} + {user_A[4]*user_B[4]} = {dot_product}")

# 4. å‘é‡é•¿åº¦ï¼ˆæ¨¡ï¼‰
length_A = np.linalg.norm(user_A)
length_B = np.linalg.norm(user_B)
print(f"\n4ï¸âƒ£ å‘é‡é•¿åº¦:")
print(f"ç”¨æˆ·Açš„å‘é‡é•¿åº¦: {length_A:.2f}")
print(f"ç”¨æˆ·Bçš„å‘é‡é•¿åº¦: {length_B:.2f}")

print(f"\nğŸ¯ å…³é”®ç†è§£:")
print(f"- å‘é‡è¿ç®— = æ•°æ®æ“ä½œ")
print(f"- ç‚¹ç§¯ = ç›¸ä¼¼åº¦æŒ‡æ ‡")
print(f"- ç‚¹ç§¯è¶Šå¤§ = è¶Šç›¸ä¼¼")
print(f"- è¿™å°±æ˜¯æ¨èç³»ç»Ÿçš„æ•°å­¦åŸºç¡€ï¼")
```

## 3ï¸âƒ£ å‘é‡åº”ç”¨ = æœºå™¨å­¦ä¹ åŸºç¡€

### ğŸ¤” æœ€é‡è¦çš„åº”ç”¨
- **ç›¸ä¼¼åº¦è®¡ç®—**ï¼šæ‰¾åˆ°ç›¸ä¼¼çš„ç”¨æˆ·æˆ–å•†å“
- **æ¨èç³»ç»Ÿ**ï¼š"å–œæ¬¢è¿™ä¸ªçš„äººä¹Ÿå–œæ¬¢..."
- **èšç±»åˆ†æ**ï¼šæŠŠç›¸ä¼¼çš„æ•°æ®ç‚¹åˆ†ç»„
- **åˆ†ç±»é¢„æµ‹**ï¼šæ ¹æ®è·ç¦»åˆ¤æ–­ç±»åˆ«

### ğŸ› ï¸ å®é™…æ¡ˆä¾‹
- **æ·˜å®æ¨è**ï¼šæ ¹æ®ä½ çš„è´­ä¹°è®°å½•æ¨èå•†å“
- **ç½‘æ˜“äº‘éŸ³ä¹**ï¼šæ ¹æ®ä½ çš„å¬æ­Œè®°å½•æ¨èæ­Œæ›²
- **Netflix**ï¼šæ ¹æ®ä½ çš„è§‚å½±è®°å½•æ¨èç”µå½±

### ğŸ’¡ æ ¸å¿ƒç†å¿µ
> "å‘é‡åº”ç”¨å°±æ˜¯ç°ä»£äº’è”ç½‘çš„æ¨èå¼•æ“"


```python
# 3ï¸âƒ£ å‘é‡åº”ç”¨ï¼šæ¨èç³»ç»Ÿ
print("=== å‘é‡åº”ç”¨ = æ¨èç³»ç»Ÿ ===")

# æ‰©å±•æ•°æ®ï¼šæ›´å¤šç”¨æˆ·çš„è¯„åˆ†
users_ratings = np.array([
    [5, 3, 4, 2, 1],  # ç”¨æˆ·Aï¼šå–œæ¬¢åŠ¨ä½œç‰‡ã€ç§‘å¹»ç‰‡
    [4, 3, 5, 2, 2],  # ç”¨æˆ·Bï¼šå–œæ¬¢åŠ¨ä½œç‰‡ã€ç§‘å¹»ç‰‡
    [1, 2, 1, 4, 5],  # ç”¨æˆ·Cï¼šå–œæ¬¢çˆ±æƒ…ç‰‡ã€ææ€–ç‰‡
    [5, 4, 4, 1, 1],  # ç”¨æˆ·Dï¼šå–œæ¬¢åŠ¨ä½œç‰‡ã€å–œå‰§ç‰‡ã€ç§‘å¹»ç‰‡
    [2, 1, 2, 5, 4],  # ç”¨æˆ·Eï¼šå–œæ¬¢çˆ±æƒ…ç‰‡ã€ææ€–ç‰‡
])

user_names = ['ç”¨æˆ·A', 'ç”¨æˆ·B', 'ç”¨æˆ·C', 'ç”¨æˆ·D', 'ç”¨æˆ·E']
movies = ['åŠ¨ä½œç‰‡', 'å–œå‰§ç‰‡', 'ç§‘å¹»ç‰‡', 'çˆ±æƒ…ç‰‡', 'ææ€–ç‰‡']

print("ç”¨æˆ·è¯„åˆ†æ•°æ®:")
for i, name in enumerate(user_names):
    print(f"{name}: {users_ratings[i]}")

# è®¡ç®—ç”¨æˆ·Aä¸å…¶ä»–ç”¨æˆ·çš„ç›¸ä¼¼åº¦
target_user = 0  # ç”¨æˆ·A
print(f"\nğŸ¯ ä¸º{user_names[target_user]}å¯»æ‰¾ç›¸ä¼¼ç”¨æˆ·:")

similarities = []
for i in range(len(users_ratings)):
    if i != target_user:
        # è®¡ç®—ä½™å¼¦ç›¸ä¼¼åº¦ï¼ˆæ›´å‡†ç¡®çš„ç›¸ä¼¼åº¦æŒ‡æ ‡ï¼‰
        sim = cosine_similarity([users_ratings[target_user]], [users_ratings[i]])[0][0]
        similarities.append((user_names[i], sim))
        print(f"{user_names[target_user]} vs {user_names[i]}: ç›¸ä¼¼åº¦ = {sim:.3f}")

# æ‰¾åˆ°æœ€ç›¸ä¼¼çš„ç”¨æˆ·
most_similar = max(similarities, key=lambda x: x[1])
print(f"\nğŸ† æœ€ç›¸ä¼¼çš„ç”¨æˆ·: {most_similar[0]} (ç›¸ä¼¼åº¦: {most_similar[1]:.3f})")

# æ¨èé€»è¾‘
print(f"\nğŸ’¡ æ¨èç³»ç»Ÿé€»è¾‘:")
print(f"1. æ‰¾åˆ°ä¸{user_names[target_user]}æœ€ç›¸ä¼¼çš„ç”¨æˆ·ï¼š{most_similar[0]}")
print(f"2. æ¨è{most_similar[0]}å–œæ¬¢ä½†{user_names[target_user]}è¿˜æ²¡é«˜è¯„åˆ†çš„ç”µå½±")
print(f"3. è¿™å°±æ˜¯'å–œæ¬¢è¿™ä¸ªçš„äººä¹Ÿå–œæ¬¢...'çš„åŸç†ï¼")

print(f"\nğŸ¯ å…³é”®ç†è§£:")
print(f"- å‘é‡åº”ç”¨ = æ¨èç³»ç»Ÿ")
print(f"- ç›¸ä¼¼åº¦ = æ¨èçš„åŸºç¡€")
print(f"- è¿™å°±æ˜¯æ·˜å®ã€ç½‘æ˜“äº‘ã€Netflixçš„æ ¸å¿ƒç®—æ³•ï¼")
```
